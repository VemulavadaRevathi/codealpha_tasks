# Exploratory Data Analysis (EDA)

In this section, we explore the scraped quotes dataset to identify patterns, trends, and insights.


```python
import pandas as pd

df = pd.read_csv("quotes.csv")

# Basic exploration
print(df.shape)          # rows & columns
print(df.columns)        # column names
print(df.info())         # datatypes
print(df.describe())     # summary stats (if numeric)
df.head()                # first 5 rows
```

- The dataset contains 100 quotes across 2 columns: Quote and Author.
- Each quote is a string, and author names are categorical data.


```python
df.duplicated().sum()   # count duplicates
df.isnull().sum()       # missing values
```

```python
df = df.drop_duplicates()
```

```python
import matplotlib.pyplot as plt

# Top authors
df['Author'].value_counts().head(10).plot(kind='bar', color="skyblue", figsize=(8,5))
plt.title("Top 10 Authors by Number of Quotes")
plt.xlabel("Author")
plt.ylabel("Number of Quotes")
plt.xticks(rotation=45)
plt.show()
```

Albert Einstein is the most quoted author in the dataset, with nearly double the quotes compared to J.K. Rowling.


```python
from wordcloud import WordCloud

all_text = " ".join(df['Quote'])
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(all_text)

plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Most Frequent Words in Quotes")
plt.show()
```

### Insights
- Einstein dominates the dataset with the most quotes.
- Common words like *life*, *truth*, and *love* show that the dataset focuses heavily on motivational and philosophical themes.
