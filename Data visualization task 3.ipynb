```python
import pandas as pd
df = pd.read_csv("quotes.csv")
print(df.head())
```

### Question: Who are the top 10 quoted authors?

The bar chart below shows the top 10 authors in the dataset ranked by the number of quotes.  
From the visualization, we can quickly see which authors dominate the collection of quotes.

### Insight:
Albert Einstein has the highest number of quotes, followed by J.K. Rowling and Jane Austen.  
This indicates that philosophical and literary figures are the most frequently cited in the dataset.



```python
import matplotlib.pyplot as plt

author_counts = df['Author'].value_counts().head(10)
author_counts.plot(kind='bar', figsize=(8,5))

plt.title("Top 10 Authors by Number of Quotes")
plt.xlabel("Authors")
plt.ylabel("Number of Quotes")
plt.xticks(rotation=45)
plt.show()
```

Top 5 Authors by Quote Distribution

The following pie chart shows the distribution of quotes among the top 5 authors in the dataset:



```python
df['Author'].value_counts().head(5).plot.pie(autopct="%1.1f%%", figsize=(6,6))
plt.title("Quotes Distribution (Top 5 Authors)")
plt.show()
```

### Word Cloud of Most Common Words in Quotes

The following Word Cloud visualizes the most frequently used words across all quotes in the dataset:


```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

all_text = " ".join(df['Quote'])
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(all_text)

plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Most Common Words in Quotes")
plt.show()
```

```python
# Data Visualization Portfolio Template
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import plotly.express as px
df = pd.read_csv('quotes.csv')
print(df.head())

author_counts = df['Author'].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=author_counts.values, y=author_counts.index, color='skyblue')
plt.title('Top 10 Authors by Number of Quotes')
plt.xlabel('Number of Quotes')
plt.ylabel('Authors')
plt.show()

df['Author'].value_counts().head(5).plot.pie(autopct='%1.1f%%', figsize=(6,6))
plt.title('Quotes Distribution (Top 5 Authors)')
plt.ylabel('')
plt.show()

all_text = ' '.join(df['Quote'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_text)
plt.figure(figsize=(12,6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Most Common Words in Quotes')
plt.show()

fig = px.bar(author_counts, x=author_counts.index, y=author_counts.values,
             labels={'x':'Author','y':'Number of Quotes'},
             title='Interactive Top 10 Authors Bar Chart')
fig.show()

numeric_df = df.select_dtypes(include='number')
if not numeric_df.empty:
    corr = numeric_df.corr()
    plt.figure(figsize=(10,8))
    sns.heatmap(corr, annot=True, cmap='coolwarm')
    plt.title('Correlation Heatmap')
    plt.show()
else:
    print("No numeric columns to generate correlation heatmap.")
```
